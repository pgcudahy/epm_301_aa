# Load data and libraries

```r
library(tidyverse)
```

```r
library(readstata13)
```

```r
outbreak_df <- read.dta13("../data/EPM_301_Version4_complete_checked.dta")
```

```r
head(outbreak_df)
```

```r
outbreak_clean_df <- outbreak_df %>%
    rename(diarrhea = V52) %>%
    rename(headache = V53) %>%
    rename(blood_in_stool = V54) %>%
    rename(fever = V55) %>%
    rename(stomach_cramps = V56) %>%
    rename(nausea = V57) %>%
    rename(muscle_aches = V58) %>%
    mutate(b_3_experienced_ilness = factor(b_3_experienced_ilness, labels=c("No", "Yes"))) %>%
    mutate(b_6_other_symptoms = factor(b_6_other_symptoms, labels=c("No", "Yes"))) %>%
    mutate(b_8_recovered = factor(b_8_recovered, labels=c("No", "Yes"))) %>%
    mutate(b_9_contacted_provider = factor(b_9_contacted_provider, labels=c("No"))) %>%
    mutate(b_10_stool_sample = factor(b_10_stool_sample, labels=c(NA))) %>%
    mutate(b_11_blood_sample = factor(b_11_blood_sample, labels=c(NA))) %>%
    mutate(b_12_diagnosis = factor(b_12_diagnosis, labels=c(NA))) %>%
    mutate(c_13_cheese = factor(c_13_cheese, labels=c("No", "Yes"))) %>%
    mutate(c_13_choc_mousse = factor(c_13_choc_mousse, labels=c("No", "Yes"))) %>%
    mutate(c_13_celery = factor(c_13_celery, labels=c("No", "Yes"))) %>%
    mutate(c_13_fruit_salad = factor(c_13_fruit_salad, labels=c("No", "Yes"))) %>%
    mutate(c_13_artichoke_dip = factor(c_13_artichoke_dip, labels=c("No", "Yes"))) %>%
    mutate(c_13_fruit_salad = factor(c_13_fruit_salad, labels=c("No", "Yes"))) %>%
    mutate(c_13_blueberries = factor(c_13_blueberries, labels=c("No", "Yes"))) %>%
    mutate(c_13_chicken_salad = factor(c_13_chicken_salad, labels=c("No", "Yes"))) %>%
    mutate(c_13_bread = factor(c_13_bread, labels=c("No", "Yes"))) %>%
    mutate(c_13_mushroom_risotto = factor(c_13_mushroom_risotto, labels=c("No", "Yes"))) %>%
    mutate(c_13_tap_water = factor(c_13_tap_water, labels=c("No", "Yes"))) %>%
    mutate(c_13_cheese_tomato = factor(c_13_cheese_tomato, labels=c("No", "Yes"))) %>%
    mutate(c_13_tea = factor(c_13_tea, labels=c("No", "Yes"))) %>%
    mutate(c_13_roasted_vegetables = factor(c_13_roasted_vegetables, labels=c("No", "Yes"))) %>%
    mutate(c_13_coffee = factor(c_13_coffee, labels=c("No", "Yes"))) %>%
    mutate(c_13_cheesecake = factor(c_13_cheesecake, labels=c("No", "Yes"))) %>%
    mutate(c_13_red_wine = factor(c_13_red_wine, labels=c("No", "Yes"))) %>%
    mutate(c_14_brought_items = factor(c_14_brought_items, labels=c("No", "Yes"))) %>%
    mutate(d_15_secondary_cases = factor(d_15_secondary_cases, labels=c("No"))) %>%
    mutate(d_17_recent_travel = factor(d_17_recent_travel, labels=c("No")))
```
